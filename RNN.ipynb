```python
nlp = spacy.load('en_core_web_sm')
```

```python
text="I am a vary very GoOOd Boy"
def clean_and_lemmitize(text):
  doc = nlp(text)

  tokens = [
      token.lemma_.lower()
      for token in doc
      if not token.is_stop and not token.is_punct
  ]
  return " ".join(tokens)

clean_and_lemmitize(text)
```
